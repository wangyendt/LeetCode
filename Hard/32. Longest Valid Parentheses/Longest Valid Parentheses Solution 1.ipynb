```python
class Solution:
    def longestValidParentheses(self, s: str) -> int:
        if s == '':
            return 0
        stack = []
        validLen = [0] * len(s)
        for i in range(len(s)):
            if '(' == s[i]:
                stack.append(i)
                validLen[i] = 0
            else:
                if stack:
                    pos = stack.pop()
                    validLen[i] = validLen[pos-1]+i-pos+1
                else:
                    validLen[i] = 0
        return max(validLen)
```

```python
so = Solution()
print(so.longestValidParentheses('()'), 2)
print(so.longestValidParentheses('(()'), 2)
print(so.longestValidParentheses('()())'), 4)
print(so.longestValidParentheses('()(()'), 2)
print(so.longestValidParentheses("())()()((()"), 4)
print(so.longestValidParentheses("()((())()"), 6)
print(so.longestValidParentheses("(((((((())()"), 6)
print(so.longestValidParentheses("()))((())()"), 6)
print(so.longestValidParentheses("(())()(()"), 6)
print(so.longestValidParentheses("((()()(()((()"), 4)
print(so.longestValidParentheses(test))
```
