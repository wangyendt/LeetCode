```python
class Solution:
    def solveNQueens(self, n: int) -> list(list()):
        ones = (1<<n)-1
        postions = [-1]*n
        ret = []
        def solve(mask, row):
            if row > n - 1 or mask[row] == 0:
                return
            else:
                while mask[row]:
                    nextTryPos = 0 # position from right to left
                    nextTryCopy = nextTry = mask[row] & -mask[row]
                    while nextTryCopy:
                        nextTryPos += 1
                        nextTryCopy >>= 1
                    postions[row] = nextTryPos
                    nextMask = mask[:]
                    nextMask[row] = 0
                    for r in range(n):
                        if r != row:
                            dis = abs(r - row)
                            localMask = ~nextTry
                            if nextTryPos-1-dis >= 0:
                                localMask &= ~(1<<(nextTryPos-1-dis))
                            if nextTryPos-1+dis < n:
                                localMask &= ~(1<<(nextTryPos-1+dis))
                            nextMask[r] &= localMask
                    
                    if row == n - 1:
                        strRet = []
                        strRow = '.'*n
                        for p in postions:
                            strRet.append('.'*(n-p)+'Q'+'.'*(p-1))
#                             print('.'*(n-p)+'Q'+'.'*(p-1))
#                         print('------------')
                        ret.append(strRet)
                    solve(nextMask, row+1)
                    mask[row] &= ~nextTry
        solve([ones for _ in range(n)], 0)
        return ret
```

```python
so = Solution()
print(so.solveNQueens(8))
```
