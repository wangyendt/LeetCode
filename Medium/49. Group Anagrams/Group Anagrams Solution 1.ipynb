```python
class Solution:
    def groupAnagrams(self, strs: list) -> list(list()):
        ret = []
        retEmpty = []
        letters = 'abcdefghijklmnopqrstuvwxyz'
        lettersDict = dict(zip(letters, range(26)))
        sortedList = [[] for _ in range(26)]
        for s in strs:
            if not s:
                retEmpty.append(s)
                continue
            t = [lettersDict[r] for r in s]
            t.sort()
            str_t = ''.join(letters[r] for r in t)
            bBreak = False
            for sl in sortedList[lettersDict[str_t[-1]]]:
                for k in sl:
                    if str_t == sl[k]:
                        ret[k].append(s)
                        bBreak = True
                        break
                if bBreak:
                    break
            else:
                ret.append([s])
                sortedList[lettersDict[str_t[-1]]].append({len(ret)-1: str_t})
        if retEmpty:
            ret.append(retEmpty)
        for i, s in enumerate(sortedList):
            print(letters[i],':',len(s))
        return ret
```
