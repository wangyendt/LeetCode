```python
class Solution:
    def threeSum(self, nums):
        ret = []
        nums.sort()
        for i in range(len(nums)):
            if i > 0:
                if nums[i] == nums[i-1]:
                    continue
            target = nums[i]
            rawList = nums[i+1:]
            composentList = [-target - j for j in rawList]
            pairs = self.findSameEntriesInTwoLists(rawList, composentList)
            if pairs != None:
                rubbish = []
                for m, n in pairs:
                    if not min(m,n) in rubbish:
                        ret.append([target, m, n])
                        rubbish.append(min(m,n))
        return ret
            
    def findSameEntriesInTwoLists(self, list1, list2):
        dict1 = {}
        cnt1, cnt2 = 0, 0
        for l1 in list1:
            dict1[l1] = cnt1
            cnt1 += 1
        rubbish = []
        for l2 in list2:
            if dict1.get(l2) is not None and dict1.get(l2) != cnt2 and (not cnt2 in rubbish):
                rubbish.append(dict1.get(l2))
                rubbish.append(cnt2)
                yield list1[cnt2], list2[cnt2]
            cnt2 += 1
```
