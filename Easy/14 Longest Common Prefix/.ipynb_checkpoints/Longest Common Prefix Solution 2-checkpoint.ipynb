```python
class Solution:
    def longestCommonPrefix(self, strs):
        self.strs = strs
        self.minSize = 10000
        return self.recursive()
        
    def recursive(self):
        l = len(self.strs)
        cnt = 0
        if l == 1:
            return self.strs[0]
        for i in range(l // 2):
            tmp = self.findCommonPrefixOfPairs(self.strs[2*i][:self.minSize], self.strs[2*i+1][:self.minSize])
            if tmp == '':
                return ''
            else:
                self.minSize = min(self.minSize, len(tmp))
                self.strs[cnt] = tmp
                cnt += 1
        if l % 2 == 1:
            self.strs[cnt] = self.strs[-1]
            self.strs = self.strs[:cnt]
        else:
            self.strs = self.strs[:cnt]
        print(len(self.strs))
        return self.recursive()
        
            
    def findCommonPrefixOfPairs(self, str1, str2):
        retStr = ''
        for i in range(min(len(str1),len(str2))):
            if str1[i] == str2[i]:
                retStr += str1[i]
            else:
                break
        return retStr
            
```
