```python
class Solution:
    def countPrimes(self, n: int) -> int:
        allPrimes = []
        def isPrime(m):       
            p = allPrimes[-1]+1 if allPrimes else 2
            while p * p <= m:
                if not m % p:
                    return False
                p += 1
            return True
        primeCnt = 0
        for i in range(2, n):
            print(allPrimes)
            for j in allPrimes:
                if not i % j:
                    break
            else:
                if isPrime(i):
                    allPrimes.append(i)
                    primeCnt += 1
        return primeCnt
```

```python
so = Solution()
print(so.countPrimes(10), 4)
print(so.countPrimes(100), 25)
print(so.countPrimes(1000))
# print(so.countPrimes(499979))
```
