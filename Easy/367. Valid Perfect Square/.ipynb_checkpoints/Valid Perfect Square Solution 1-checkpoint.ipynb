```python
class Solution:
    def isPerfectSquare(self, num: int) -> bool:
        l, r = 0, num
        if num == 1 or num == 0:
            return True
        while r-l > 1:
            m = (l+r)//2
            if m**2 > num:
                r = m
            elif m**2 == num:
                return True
            else:
                l = m
        for i in range(l, r+1):
            if i**2 == num:
                return True
        return False
```

```python
so = Solution()
[so.isPerfectSquare(i) for i in range(200000)]
```
