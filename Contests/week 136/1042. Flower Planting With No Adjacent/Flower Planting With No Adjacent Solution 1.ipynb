```python
class Solution:
    def gardenNoAdj(self, N: int, paths: list(list())) -> list:
        connDict = {}
        ret = [1] * N
        for p in paths:
            s, t = p
            s, t = (t, s) if t < s else (s, t)
            connS, connT = connDict.get(s), connDict.get(t)
            if not connS and not connT:
                ret[s-1] = 1
                ret[t-1] = 2
                connDict[s] = [t]
                connDict[t] = [s]
            elif not connS and connT:
                ret[s-1] = 1 if ret[t-1] != 1 else 2
                connDict[s] = [t]
                connDict[t].append(s)
            elif connS and not connT:
                ret[t-1] = 1 if ret[s-1] != 1 else 2
                connDict[s].append(t)
                connDict[t] = [s]
            else:
                if ret[s-1] == ret[t-1]:
                    stack = [1,2,3,4]
                    for r in connT:
                        if ret[r-1] in stack:
                            stack.pop(stack.index(ret[r-1]))
                    if ret[s-1] in stack:
                        stack.pop(stack.index(ret[s-1]))
                    if not stack:
                        ret[t-1] = stack[0]
                    else:
                        stack = [1,2,3,4]
                        for r in connS:
                            if ret[r-1] in stack:
                                stack.pop(stack.index(ret[r-1]))
                        if ret[t-1] in stack:
                            stack.pop(stack.index(ret[t-1]))
                        ret[s-1] = stack[0]
                connDict[s].append(t)
                connDict[t].append(s)
        return ret
```
